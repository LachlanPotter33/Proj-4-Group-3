```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame

# Specify the file path
file_path = "../Data/prosperLoanData.csv"
```

```python
# Read the CSV file into a Pandas DataFrame
Uncleaned_df = pd.read_csv(file_path)

# Review the DataFrame

Uncleaned_df.head()
```

```python
#Creating Credit Score ( Lowest + Highest credit score /2 )
Uncleaned_df['Average Credit Score'] = (Uncleaned_df['CreditScoreRangeLower'] + Uncleaned_df['CreditScoreRangeUpper']) / 2
```

```python
#types of columns 
specific_columns = ['LoanStatus','Term','ProsperRating (numeric)','Average Credit Score','EmploymentStatus','IsBorrowerHomeowner','StatedMonthlyIncome',
'BorrowerRate','EmploymentStatusDuration','LoanMonthsSinceOrigination','MonthlyLoanPayment','RevolvingCreditBalance','LoanOriginalAmount'  ] 
```

```python
specific_columns_uncleaned_df = Uncleaned_df[specific_columns]
specific_columns_uncleaned_df.head(10)
```

```python
specific_columns_uncleaned_df = specific_columns_uncleaned_df.dropna()
```

```python
#changing data that is not a value to be NA
specific_columns_uncleaned_df.replace("Not displayed", pd.NA, inplace=True)
```

```python
#changing data in Loan Status to be numbers
LoanStatus_unique_values = specific_columns_uncleaned_df['LoanStatus'].unique()
#good load - 0 
# at risk -1
#Bad debts - 2 

mapping = {
    'Completed': 0,
    'Current': 0,
    'FinalPaymentInProgress': 0,
    'Past Due (1-15 days)': 1,
    'Defaulted': 2,
    'Chargedoff':2,
    'Past Due (16-30 days)':1,
    'Past Due (61-90 days)':1,
    'Past Due (31-60 days)':1,
    'Past Due (91-120 days)': 1,
    'Past Due (>120 days)': 1
}
# Replace the categorical values with numerical values
specific_columns_uncleaned_df['LoanStatus'] = specific_columns_uncleaned_df['LoanStatus'].replace(mapping)
specific_columns_uncleaned_df
```

```python
#changing data in Employment Status to be numbers
EmploymentStatus_unique_values = specific_columns_uncleaned_df['LoanStatus'].unique()
mapping = {
    'Self-employed': 1,
    'Employed': 2,
    'Not available': 3,
    'Full-time': 4,
    'Other': 5,
    'Not employed': 6,
    'Part-time': 7,
    'Retired': 8
}
# Replace the categorical values with numerical values
specific_columns_uncleaned_df['EmploymentStatus'] = specific_columns_uncleaned_df['EmploymentStatus'].replace(mapping)
specific_columns_uncleaned_df
```

```python
specific_columns_uncleaned_df = pd.DataFrame(specific_columns_uncleaned_df, columns=['LoanStatus','Term','ProsperRating (numeric)','Average Credit Score','EmploymentStatus','StatedMonthlyIncome',
'BorrowerRate','EmploymentStatusDuration','LoanMonthsSinceOrigination','MonthlyLoanPayment','RevolvingCreditBalance','LoanOriginalAmount'])
```

```python
specific_columns_uncleaned_df['IsBorrowerHomeowner'] = specific_columns_uncleaned_df['IsBorrowerHomeowner'].map({True: 1, False: 0})
```

```python
specific_columns_uncleaned_df.head()
```

```python
# Separate the data into labels and features

y = specific_columns_uncleaned_df['LoanStatus']

# Separate the X variable, the features
X = specific_columns_uncleaned_df.drop(columns=['LoanStatus'])
```

```python
# Review the y variable Series
#0 means healthy 1means high risk 2means bad loan
y.head()
# Review the X variable DataFrame
X.head()
```

```python
# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
```

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression


# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier

# Fit the model using training data
classifier.fit(X_train, y_train)
```

```python
# Score the model
print(f"Loan Score: {classifier.score(X_train, y_train)}")
print(f"Loan Score: {classifier.score(X_test, y_test)}")
```

```python
# Make a prediction using the testing data
predictions = classifier.predict(X_test)

print(len(predictions))
```

```python
# Generate a confusion matrix for the model
cm_df=confusion_matrix(y_test, predictions)
display(cm_df)
```

```python
# Print the classification report for the model
target_names = ["0-Healthy", "1-High Risk", "2-Bad Loan"]
print(classification_report(y_test, predictions, target_names=target_names))
```

```python
columns_to_scale = ['LoanStatus' ,
                    'Term',
                     'ProsperRating (numeric)',
                     'Average Credit Score',
                     'EmploymentStatus',
                     'IsBorrowerHomeowner',
                     'StatedMonthlyIncome',
                     'BorrowerRate',
                     'EmploymentStatusDuration',
                     'LoanMonthsSinceOrigination',
                     'MonthlyLoanPayment',
                     'RevolvingCreditBalance',
                     'LoanOriginalAmount']
```

```python
specific_columns_uncleaned_df = specific_columns_uncleaned_df.dropna(subset=columns_to_scale)
```
