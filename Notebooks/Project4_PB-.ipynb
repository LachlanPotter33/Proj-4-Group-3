```python
# Importing essential libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
# Main_path = f'{os.getcwd()}'
Prosper_df = pd.read_csv('..\Data\prosperLoanData.csv')
Prosper_df.head()
```

```python
# Creating Credit Score ( Lowest + Highest credit score /2 )
Prosper_df['Average Credit Score'] = (Prosper_df['CreditScoreRangeLower'] + Prosper_df['CreditScoreRangeUpper']) / 2
```

```python
# Filling null values
Prosper_df.interpolate(method='linear', inplace=True)
```

```python
# Making data frame with specific column 
specific_columns_df = ['Term','LoanStatus','ProsperRating (numeric)','Average Credit Score','EmploymentStatus','IsBorrowerHomeowner','IncomeRange','StatedMonthlyIncome','BorrowerRate','EmploymentStatusDuration','CurrentDelinquencies','LoanMonthsSinceOrigination','LoanOriginalAmount','TotalProsperPaymentsBilled', 'OnTimeProsperPayments','PublicRecordsLast12Months','PublicRecordsLast10Years']
```

```python
# Select specific columns
specific_columnsss = Prosper_df[specific_columns_df]
```

```python
# Check for missing values
Prosper_df.isnull().sum()
```

```python
# Info about the DataFrame
specific_columnsss.info()
```

```python
# Convert IsBorrowerHomeowner to float
specific_columnsss['IsBorrowerHomeowner'] = specific_columnsss['IsBorrowerHomeowner'].astype(float)
# Unique values in IsBorrowerHomeowner
Ownership_unique_values = specific_columnsss['IsBorrowerHomeowner'].unique()
# Map IsBorrowerHomeowner values
mapping = {
    'True': 1,
    'False': 0,
    }
specific_columnsss['IsBorrowerHomeowner'] = specific_columnsss['IsBorrowerHomeowner'].replace(mapping)
specific_columnsss
```

```python
# Unique values in EmploymentStatus
EmploymentStatus_unique_values = specific_columnsss['EmploymentStatus'].unique()
# Map EmploymentStatus values
mapping = {
    'Self-employed': 1,
    'Employed': 1,
    'Not available': 0,
    'Full-time': 1,
    'Other': 0,
    'Not employed': 0,
    'Part-time': 2,
    'Retired': 3,
    'NaN': 0,
}
specific_columnsss['EmploymentStatus'] = specific_columnsss['EmploymentStatus'].replace(mapping)
specific_columnsss
```

```python
# Drop missing values
specific_columnsss = specific_columnsss.dropna()
```

```python
# Unique values in IncomeRange
IncomeRange_Unique = specific_columnsss['IncomeRange'].unique()
# Map IncomeRange values

mapping = {
    '$0': 0,
    '$1-24,999': 1,
    '$25,000-49,999': 1,
    '$50,000-74,999': 2,
    '$75,000-99,999': 2,
    '$100,000+': 3,
    'Not employed': 0,
    'Not displayed': 0
}
specific_columnsss['IncomeRange'] = specific_columnsss['IncomeRange'].replace(mapping)
specific_columnsss
```

```python
# Unique values in LoanStatus
IncomeRange_Unique = specific_columnsss['LoanStatus'].unique()
# Map LoanStatus values
mapping = {    
    'Current': 0,
    'Completed': 0,
    'FinalPaymentInProgress': 0,
    'Past Due (1-15 days)': 1,
    'Past Due (16-30 days)': 1,
    'Past Due (31-60 days)': 1,
    'Past Due (61-90 days)': 2,
    'Past Due (91-120 days)': 2,
    'Past Due (>120 days)': 2,
    'Chargedoff':2,
    'Defaulted':2,
    'Cancelled': 2
       
}
specific_columnsss['LoanStatus'] = specific_columnsss['LoanStatus'].replace(mapping)
specific_columnsss
```

```python
# Unique loan statuses
loan_statuses = specific_columnsss['LoanStatus'].unique()

# Parameters for pair plot
height = 2
aspect = 1.2

# Filter data for each loan status
for status in loan_statuses:

    filtered_df = specific_columnsss[specific_columnsss['LoanStatus'] == status]

# Filter columns for pair plot
filtered_df = specific_columnsss[['LoanOriginalAmount', 'TotalProsperPaymentsBilled', 'OnTimeProsperPayments', 'CurrentDelinquencies']]
```

```python
# Import KMeans
from sklearn.cluster import KMeans

# Filter relevant columns for clustering
filtered_columns = ['LoanOriginalAmount', 'TotalProsperPaymentsBilled', 'OnTimeProsperPayments', 'CurrentDelinquencies']
filtered_data = specific_columnsss[filtered_columns].dropna()

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(filtered_data)
specific_columnsss['Cluster'] = kmeans.labels_

# Scatter plot for clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='LoanOriginalAmount', y='TotalProsperPaymentsBilled', hue='Cluster', data=specific_columnsss, palette='viridis')
plt.title('Clusters of Loans')
plt.xlabel('Loan Original Amount')
plt.ylabel('Total Prosper Payments Billed')
plt.show()
```
